# Data
## To solve this problem we will need the following data:
* Postal codes, boroughs and neighbourhoods of Toronto, available here: <a href="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"> BOROUGHS </a>
* Latitude and longitude coordinates of those suburbs, available here: <a href="https://cocl.us/Geospatial_data"> COORDINATES </a>
Venue data, particularly data related to coffee shops (Foursqare API)

The Wikipedia page contains a list of postal codes of Toronto with a total of 180 of them. Python and beautifulsoup packages will be used to help with web scraping the data. Coordinates will be taken from 'Geospatial' data. Foursqare API has one of the largest database of more than 105 million places. It will help us to acquire data of Toronto neighbourhoods' cofee shops.

# Methodology
The data containing boroughs and neighbourhoods of Toronto was scraped from Wikipedia. After that it was cleaned and additional data (location) was obtained from https://cocl.us/Geospatial_data. Both informations were merged into new table which consisted of Postal Code, Borough, Neighbourhood, Latitude and Longitude.<br>
After calling credentials the search query: <strong> "coffee"</strong> was passed using API Foursquare scope. The result was transformed from json to pandas data frame. After cleaning it, new data was created and included: name, latitude and longitude of coffee shops in Toronto area. <br>
The data was clustered (using <strong> KNN </strong> clustering algorithm) into 6 groups and a map was created (using <strong> Folium </strong> properties). To visualize and indicate each cluster different colors were used (thanks to <strong> matplotlib </strong> rainbow libraries).

# Results



You can clearly see which area of Toronto map has the highest density of coffee shops since they are assigned into <strong> 6 clusters </strong>. The stakeholder might take into consideration the areas with not many shops around to eliminate potential competition or places next to certain infrastructure like university or shopping malls.
